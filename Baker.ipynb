```python
site = pd.read_csv('site_metadata-7fc535965d0c074cea0be6786fc9518e.csv')
print(site.loc[0,'PLANT_NAME'])
engine = pd.read_csv('engine_metadata-fad08922e295794745ba1cd70379140f.csv')
data_2 = pd.read_csv('data_1.csv')
print(site.head(10))
print(engine.head(10))
print(site.loc[:,'CUSTOMER_NAME'])
```

```python
df = pd.DataFrame()
df['Customers'] = engine.loc[:,'CUSTOMER_NAME']
df['Plant name'] = engine.loc[:,'PLANT_NAME']
df['Latitude'] = site.loc[:,'LATITUDE']
df['Longitude'] = site.loc[:,'LONGITUDE']
df['Elevation'] = site.loc[:,'ELEVATION']
df['Engine ID'] = engine.loc[:,'ENGINE_ID']
df['File ID'] = engine.loc[:,'FILE_ID']
print(df.head(10))
```

```python
plant_dict = dict(zip(site.PLANT_NAME, site.FUEL_LHV))
```

```python
plantlist = []
plantlist_avg = []
plantcounter = 0
co2list = []
speedlist = []
powerlist = []
fuellist = []
hourslist = []
engine_list = df['Engine ID'].tolist()
file_list = df['File ID'].tolist()
index = 0
therm_total = 0
co2_total = 0
speed_total = 0
power_total = 0
fuel_total = 0
hours_total = 0
plantcounter = 0
co2counter = 0
speedcounter = 0
powercounter = 0
fuel_counter = 0
hours_counter = 0
for i,name in enumerate(engine_list):
    letters = list(name)
    num = letters[-1]
    fuel_lhv = plant_dict[df.loc[i,'Plant name']]
    data_name = 'data_' + str(i+1) + '.csv'
    cur_data = pd.read_csv(data_name)
    #solving for operation hours
    hours_list = cur_data['CMP_SPEED'].tolist()
    hours = 0
    num_valid = 0
    speed = 0
    for f,value in enumerate(hours_list):
        if f != len(hours_list) - 1:
            if value != 0 and hours_list[f] != 0:
                hours += 1
            if value != 0:
                speed += value
                num_valid += 1
    speed_avg = speed / num_valid
    
    #solving for co2 of data set
    co2_avg = cur_data['CO2'].mean()
    
    #gettting thermal efficiency
    power = cur_data['POWER'].mean()
    power_avg = power
    fuel_flow = cur_data['FUEL_FLOW'].mean()
    fuel_avg = fuel_flow
    therm = (100 * power / (fuel_flow * fuel_lhv))/(0.45 * 1.055085)
    if int(num) == 1:
        therm_total = therm
        co2_total = co2_avg
        speed_total = speed_avg
        power_total = power_avg
        fuel_total = fuel_avg
        hours_total = hours
        plantlist.append(site.loc[index,'PLANT_NAME'])
        index += 1
        plantcounter = 1
        co2counter = 1
        speedcounter = 1
        powercounter = 1
        fuelcounter = 1
        hourscounter = 1
        if i != len(engine_list) - 1:
            next_letters = engine_list[i+1]
            next_list = list(next_letters)
            if int(next_list[-1]) == 1:
                avg_co2 = co2_total / co2counter
                avg = therm_total / plantcounter
                avg_speed = speed_total / speedcounter
                avg_power = power_total / powercounter
                avg_fuel = fuel_total / fuelcounter
                avg_hours = hours_total / hourscounter
                plantlist_avg.append(avg)
                co2list.append(avg_co2)
                speedlist.append(avg_speed)
                powerlist.append(avg_power)
                fuellist.append(avg_fuel)
                hourslist.append(avg_hours)
    else: 
        if i != len(engine_list) - 1:
            therm_total += therm
            co2_total += co2_avg
            speed_total += speed_avg
            power_total += power_avg
            fuel_total += fuel_avg
            hours_total += hours
            plantcounter += 1
            co2counter += 1
            speedcounter += 1
            powercounter += 1
            fuelcounter += 1
            hourscounter += 1
            next_letters = engine_list[i+1]
            next_list = list(next_letters)
            if int(next_list[-1]) == 1:
                avg_co2 = co2_total / co2counter
                avg = therm_total / plantcounter
                avg_speed = speed_total / speedcounter
                avg_power = power_total / powercounter
                avg_fuel = fuel_total / fuelcounter
                avg_hours = hours_total / hourscounter
                plantlist_avg.append(avg)
                co2list.append(avg_co2)
                speedlist.append(avg_speed)
                powerlist.append(avg_power)
                fuellist.append(avg_fuel)
                hourslist.append(avg_hours)
        else: 
            therm_total += therm
            co2_total += co2_avg
            speed_total += speed_avg
            power_total += power_avg
            fuel_total += fuel_avg
            hours_total += hours
            plantcounter += 1
            co2counter += 1
            speedcounter += 1
            powercounter += 1
            fuelcounter += 1
            hourscounter += 1
            avg_co2 = co2_total / co2counter
            avg = therm_total / plantcounter
            avg_speed = speed_total / speedcounter
            avg_power = power_total / powercounter
            avg_fuel = fuel_total / fuelcounter
            avg_hours = hours_total / hourscounter
            plantlist_avg.append(avg)
            co2list.append(avg_co2)
            speedlist.append(avg_speed)
            powerlist.append(avg_power)
            fuellist.append(avg_fuel)
            hourslist.append(avg_hours)
            
plant_dataframe = pd.DataFrame()
plant_dataframe['Plant name'] = plantlist
plant_dataframe['Thermal efficiency averages'] = plantlist_avg 
plant_dataframe['CO2 emissions'] = co2list
plant_dataframe['Average Comp speed'] = speedlist
plant_dataframe['Average Power'] = powerlist
plant_dataframe['Average Fuel Flow'] = fuellist
plant_dataframe['Average Operating Hours'] = hourslist
print(plant_dataframe)
```
